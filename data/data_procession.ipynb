```python
import pandas as pd

# 读取数据
data_raw = pd.read_excel('raw_data_include_ref.xlsx')
data_raw.head()
```

```python
# 列清理
data = data_raw.drop(["Index", "Cathode", "Smicro", "Smeso", "Vmicro", "Vmeso", "Electrolyte", "Ref"], axis=1)
data
```

```python
# 填充缺失值
data["AML"] = data["AML"].fillna(data["AML"].median())

# 数据分布统计
statistics = data.describe(include='all')
statistics
```

```python
# 保存一份数据
data.to_csv('1_dataset_fillna.csv', index=False)
```
