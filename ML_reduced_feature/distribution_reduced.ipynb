```python
from matplotlib import pyplot as plt
import pandas as pd

# 读取数据
data = pd.read_csv("../data/dataset_reduced.csv")
data = data.drop(columns=['Anion'])
# 统计分布，绘制直方图
data.describe()

# 绘制直方图
data.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 定义变换函数
transformations = {
    'sin': np.sin,
    'cos': np.cos,
    'sqrt': np.sqrt,
    'square': np.square,
    'log': lambda x: np.log(x + 1)
}

# 对每种变换绘制直方图
for name, func in transformations.items():
    transformed_data = data.apply(func)
    transformed_data.hist(bins=30, figsize=(15, 10))
    plt.suptitle(f'Transformation: {name}')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
